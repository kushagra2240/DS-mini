building a random forest with all numerical parameters

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

y = clean_data.Price

#predictors = ['Rooms','BuildingArea','Lattitude','Longtitude','Landsize']
predictors = num_data.columns
X = clean_data[predictors]
trainX, valX, train_y, val_y = train_test_split(X,y,random_state=0)

model = RandomForestRegressor()
model.fit(trainX,train_y)

predictions = model.predict(valX)
print (mean_absolute_error(val_y,predictions))
fig, ax = plt.subplots()
ax.scatter(val_y, predictions, edgecolors=(0, 0, 0))
ax.plot([val_y.min(), val_y.max()], [val_y.min(), val_y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
from statistics import mean
avgPred_coords =[]

for i in range (10) :
    y = clean_data.Price
    predictors = ['Rooms','BuildingArea','Lattitude','Longtitude','Landsize']
    X = clean_data[predictors]
    trainX, valX, train_y, val_y = train_test_split(X,y,random_state=0)

    model = RandomForestRegressor()
    model.fit(trainX,train_y)

    predictions = model.predict(valX)
    print (mean_absolute_error(val_y,predictions))
    avgPred_coords.append(mean_absolute_error(val_y,predictions))

print ('average mean absolute error is ',mean(avgPred_coords))
```
